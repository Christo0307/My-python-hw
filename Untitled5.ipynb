```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import time

#以下兩行是為了解決無法pyplot.text無法顯示中文用的
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei']

r = 1.0 #半徑
a, b = (0., 0.) #原點
theta = np.arange(0, 2*np.pi, 0.1)

x = a + r * np.cos(theta)
y = b + r * np.sin(theta)

ham = input('請輸入一串文字')
egg = len(ham)
```

```python
for i in range(len(theta)):
    plt.text(x[i], y[i], ham[i%egg], fontsize=20)
plt.axis('off')
```

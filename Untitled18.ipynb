```python
%matplotlib inline

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd

mpl.rc('font', family='SimHei')
```

```python
#基尼係數-表示貧富差距
url_gini = 'https://en.wikipedia.org/wiki/List_of_countries_by_income_equality'
df_gini = pd.read_html(url_gini)[3]
```

```python
colname_gini = np.array(['Country', 'UN R/P (10%)', 'UN R/P (20%)', 'World Bank Gini (%)', 'World Bank Gini (year)', 'CIA R/P (10%)',
        'CIA R/P (year)', 'CIA Gini (%)', 'CIA Gini (year)'], dtype=object)
df_gini.columns = colname_gini
index_gini = df_gini.loc[:,'Country']
df_gini.index = index_gini
```

```python
#購買力平價，表示薪水與物價的比
url_ppp = 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(PPP)_per_capita'
df_ppp = pd.read_html(url_ppp)[3]
```

```python
colname_ppp = df_ppp.loc[0].values
df_ppp.columns = colname_ppp
index_ppp = df_ppp.loc[:,'Country/Territory']
df_ppp.index = index_ppp
```

```python
#可支配收入，表示扣掉房租等等固定開銷還剩多少錢
url_dis = 'https://en.wikipedia.org/wiki/Disposable_household_and_per_capita_income'
df_dis = pd.read_html(url_dis)[1]
```

```python
colname_dis = df_dis.loc[1].values
df_dis.columns = colname_dis
index_dis = df_dis.loc[:,'Country/Territory']
df_dis.index = index_dis
```

```python
df_dis = df_dis.iloc[2:36].loc[:,'2015 Per Capita (PPP)[3]']
```

```python
df_GueDao = pd.DataFrame(index = df_dis.index)
df_GueDao['可支配收入'] = df_dis.copy()
df_GueDao['購買力平價'] = 0
df_GueDao['基尼係數'] = 0
df_GueDao['鬼島指數'] = 0
df_GueDao = df_GueDao.drop(index='European Union')
df_GueDao = df_GueDao.drop(index='Slovak Republic')
```

```python
for i in range(df_GueDao.shape[0]):
    ham = df_GueDao.index[i]
    if (ham == 'South Korea'):
        egg = df_ppp.at['Korea, South','Int$']
    else:
        egg = df_ppp.at[ham,'Int$']
    df_GueDao.at[ham,'購買力平價'] = egg
```

```python
for i in range(df_GueDao.shape[0]):
    ham = df_GueDao.index[i]
    if (ham == 'New Zealand'):
        egg = df_gini.at[ham, 'CIA Gini (%)']
    else:
        egg = df_gini.at[ham, 'World Bank Gini (%)']
    df_GueDao.at[ham, '基尼係數'] = float(egg)
```

```python
import locale
locale.setlocale(locale.LC_NUMERIC, '')
df_GueDao['可支配收入'] = df_GueDao['可支配收入'].apply(locale.atof)
```

```python
df_GueDao['鬼島指數'] = df_GueDao['可支配收入'] + df_GueDao['購買力平價'] - df_GueDao['基尼係數']*1000
```
